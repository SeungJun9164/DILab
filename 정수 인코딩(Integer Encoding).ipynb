# dictionary 사용

```
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```
text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
```

```
text = sent_tokenize(text)
print(text)
```

```
vocab = {}
sentences = []
stop_words = set(stopwords.words('english'))

for i in text:
  sentence = word_tokenize(i)
  result = []
  for word in sentence:
    word = word.lower()
    if word not in stop_words:
      if len(word) > 2:
        result.append(word)
        if word not in vocab:
          vocab[word] = 0
        vocab[word] += 1
  sentences.append(result)
print(sentences)
```

```
print(vocab)
```

```
print(vocab["barber"])
```

```
vocab_sorted = sorted(vocab.items(), key=lambda x:x[1], reverse = True)
print(vocab_sorted)
```

```
word_to_index = {}
i = 0
for (word, frequency) in vocab_sorted:
  if frequency > 1:
    i = i+1
    word_to_index[word] = i
print(word_to_index)
```

```
vocab_size = 5
words_frequency = [w for w,c in word_to_index.items() if c >= vocab_size + 1]
for w in words_frequency:
  del word_to_index[w]
print(word_to_index)
```

# OOV : 단어 집합에 존재하지 않는 단어들

```
word_to_index['OOV'] = len(word_to_index) + 1
```

```
encoded = []
for s in sentences:
  temp = []
  for w in s:
    try:
      temp.append(word_to_index[w])
    except KeyError:
      temp.append(word_to_index['OOV'])
  encoded.append(temp)
print(encoded)
```

# Counter 사용

```
from collections import Counter
```

```
print(sentences)
```

```
words = sum(sentences, []) # 단어 집합을 만들기 위해
print(words)
```

```
vocab = Counter(words)
print(vocab)
```

```
# most_common : 상위 빈도수를 가진 주어진 수만 반환
vocab_size = 5
vocab = vocab.most_common(vocab_size)
vocab
```

```
word_to_index = {}
i = 0
for (word, frequency) in vocab:
  i = i + 1
  word_to_index[word] = i
print(word_to_index)
```

# nltk의 FreqDist 사용

```
from nltk import FreqDist
import numpy as np
```

```
# hstack : 두 배열을 왼쪽에서 오른쪽으로 붙이기(행)
# vstack : 두 배열을 위에서 아래로 붙이기(열)
vocab = FreqDist(np.hstack(sentences))
vocab
```

```
vocab_size = 5
vocab = vocab.most_common(vocab_size)
vocab
```

```
word_to_index = {word[0] : index + 1 for index, word in enumerate(vocab)}
print(word_to_index)
```
